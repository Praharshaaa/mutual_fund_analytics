```python
import pandas as pd
import numpy as np

path = r'C:\Users\PRAHARSHA\mutual_fund_analytics\data\raw\\'

nav_history = pd.read_csv(path + '02_nav_history.csv')
investor_transactions = pd.read_csv(path + '08_investor_transactions.csv')
scheme_performance = pd.read_csv(path + '07_scheme_performance.csv')

print("✅ Files loaded!")
print(f"NAV History: {nav_history.shape}")
print(f"Investor Transactions: {investor_transactions.shape}")
print(f"Scheme Performance: {scheme_performance.shape}")
```

```python
# Clean nav_history.csv
nav_history['date'] = pd.to_datetime(nav_history['date'], format='mixed', dayfirst=True)
nav_history = nav_history.sort_values(['amfi_code', 'date'])
nav_history = nav_history.drop_duplicates()
nav_history = nav_history[nav_history['nav'] > 0]

print("✅ NAV History cleaned!")
print(f"Shape: {nav_history.shape}")
print(nav_history.head())
```

```python
# Clean investor_transactions.csv
investor_transactions['transaction_date'] = pd.to_datetime(
    investor_transactions['transaction_date'], format='mixed', dayfirst=True)
investor_transactions['amount_inr'] = pd.to_numeric(
    investor_transactions['amount_inr'], errors='coerce')
investor_transactions = investor_transactions[investor_transactions['amount_inr'] > 0]
investor_transactions['transaction_type'] = investor_transactions['transaction_type'].str.strip().str.capitalize()
investor_transactions = investor_transactions.drop_duplicates()

print("✅ Investor Transactions cleaned!")
print(f"Shape: {investor_transactions.shape}")
print(investor_transactions['transaction_type'].value_counts())
```

```python
# Clean scheme_performance.csv
cols = ['return_1yr_pct', 'return_3yr_pct', 'return_5yr_pct', 
        'alpha', 'beta', 'sharpe_ratio', 'expense_ratio_pct']

for col in cols:
    scheme_performance[col] = pd.to_numeric(scheme_performance[col], errors='coerce')

# Validate expense ratio range 0.1% - 2.5%
anomalies = scheme_performance[
    (scheme_performance['expense_ratio_pct'] < 0.1) | 
    (scheme_performance['expense_ratio_pct'] > 2.5)
]
print(f"Expense ratio anomalies: {len(anomalies)}")
scheme_performance = scheme_performance.drop_duplicates()

print("✅ Scheme Performance cleaned!")
print(f"Shape: {scheme_performance.shape}")
print(scheme_performance.head(2))
```

```python
import sqlalchemy as sa
from pathlib import Path

# Save cleaned CSVs
processed_path = r'C:\Users\PRAHARSHA\mutual_fund_analytics\data\processed\\'
Path(processed_path).mkdir(parents=True, exist_ok=True)

nav_history.to_csv(processed_path + 'nav_history_clean.csv', index=False)
investor_transactions.to_csv(processed_path + 'investor_transactions_clean.csv', index=False)
scheme_performance.to_csv(processed_path + 'scheme_performance_clean.csv', index=False)

print("✅ Cleaned CSVs saved!")

# Create SQLite database
db_path = r'C:\Users\PRAHARSHA\mutual_fund_analytics\data\db\bluestock_mf.db'
Path(r'C:\Users\PRAHARSHA\mutual_fund_analytics\data\db').mkdir(parents=True, exist_ok=True)

engine = sa.create_engine(f'sqlite:///{db_path}')

nav_history.to_sql('nav_history', engine, if_exists='replace', index=False)
investor_transactions.to_sql('investor_transactions', engine, if_exists='replace', index=False)
scheme_performance.to_sql('scheme_performance', engine, if_exists='replace', index=False)

print("✅ SQLite database created!")
print(f"Database saved at: {db_path}")
```

```python
# 10 SQL Queries
import pandas as pd

# Query 1: Top 5 funds by AUM
q1 = pd.read_sql("""
    SELECT scheme_name, fund_house, aum_crore 
    FROM scheme_performance 
    ORDER BY aum_crore DESC LIMIT 5
""", engine)
print("Q1 - Top 5 funds by AUM:")
print(q1)

# Query 2: Average NAV per month
q2 = pd.read_sql("""
    SELECT strftime('%Y-%m', date) as month, 
    ROUND(AVG(nav), 2) as avg_nav
    FROM nav_history 
    GROUP BY month 
    ORDER BY month DESC LIMIT 10
""", engine)
print("\nQ2 - Average NAV per month:")
print(q2)

# Query 3: SIP vs Lumpsum vs Redemption count
q3 = pd.read_sql("""
    SELECT transaction_type, COUNT(*) as count, 
    ROUND(SUM(amount_inr),2) as total_amount
    FROM investor_transactions 
    GROUP BY transaction_type
""", engine)
print("\nQ3 - Transaction Summary:")
print(q3)
```

```python
# Query 4: Funds with expense ratio < 1%
q4 = pd.read_sql("""
    SELECT scheme_name, expense_ratio_pct, category
    FROM scheme_performance 
    WHERE expense_ratio_pct < 1.0
    ORDER BY expense_ratio_pct ASC
""", engine)
print("Q4 - Funds with expense ratio < 1%:")
print(q4)

# Query 5: Top 5 funds by 1yr return
q5 = pd.read_sql("""
    SELECT scheme_name, return_1yr_pct, return_3yr_pct
    FROM scheme_performance 
    ORDER BY return_1yr_pct DESC LIMIT 5
""", engine)
print("\nQ5 - Top 5 funds by 1yr return:")
print(q5)

# Query 6: Transactions by state
q6 = pd.read_sql("""
    SELECT state, COUNT(*) as transactions, 
    ROUND(SUM(amount_inr),2) as total_amount
    FROM investor_transactions 
    GROUP BY state 
    ORDER BY total_amount DESC LIMIT 5
""", engine)
print("\nQ6 - Top 5 states by transaction amount:")
print(q6)

# Query 7: SIP growth year over year
q7 = pd.read_sql("""
    SELECT strftime('%Y', transaction_date) as year,
    COUNT(*) as sip_count,
    ROUND(SUM(amount_inr),2) as total_sip
    FROM investor_transactions
    WHERE transaction_type = 'Sip'
    GROUP BY year ORDER BY year
""", engine)
print("\nQ7 - SIP YoY growth:")
print(q7)

# Query 8: Average Sharpe ratio by category
q8 = pd.read_sql("""
    SELECT category, ROUND(AVG(sharpe_ratio),2) as avg_sharpe
    FROM scheme_performance
    GROUP BY category ORDER BY avg_sharpe DESC
""", engine)
print("\nQ8 - Avg Sharpe by category:")
print(q8)

# Query 9: Funds with negative max drawdown > -20%
q9 = pd.read_sql("""
    SELECT scheme_name, max_drawdown_pct
    FROM scheme_performance
    WHERE max_drawdown_pct < -20
    ORDER BY max_drawdown_pct ASC
""", engine)
print("\nQ9 - Funds with drawdown > -20%:")
print(q9)

# Query 10: Top investors by transaction amount
q10 = pd.read_sql("""
    SELECT investor_id, COUNT(*) as transactions,
    ROUND(SUM(amount_inr),2) as total_invested
    FROM investor_transactions
    GROUP BY investor_id
    ORDER BY total_invested DESC LIMIT 5
""", engine)
print("\nQ10 - Top 5 investors:")
print(q10)
```

```python
# Save queries to sql/queries.sql file
queries = """
-- Q1: Top 5 funds by AUM
SELECT scheme_name, fund_house, aum_crore 
FROM scheme_performance 
ORDER BY aum_crore DESC LIMIT 5;

-- Q2: Average NAV per month
SELECT strftime('%Y-%m', date) as month, ROUND(AVG(nav), 2) as avg_nav
FROM nav_history GROUP BY month ORDER BY month DESC LIMIT 10;

-- Q3: Transaction Summary
SELECT transaction_type, COUNT(*) as count, ROUND(SUM(amount_inr),2) as total_amount
FROM investor_transactions GROUP BY transaction_type;

-- Q4: Funds with expense ratio < 1%
SELECT scheme_name, expense_ratio_pct FROM scheme_performance 
WHERE expense_ratio_pct < 1.0 ORDER BY expense_ratio_pct ASC;

-- Q5: Top 5 funds by 1yr return
SELECT scheme_name, return_1yr_pct FROM scheme_performance 
ORDER BY return_1yr_pct DESC LIMIT 5;

-- Q6: Top 5 states by transaction amount
SELECT state, COUNT(*) as transactions, ROUND(SUM(amount_inr),2) as total_amount
FROM investor_transactions GROUP BY state ORDER BY total_amount DESC LIMIT 5;

-- Q7: SIP YoY growth
SELECT strftime('%Y', transaction_date) as year, COUNT(*) as sip_count
FROM investor_transactions WHERE transaction_type = 'Sip' GROUP BY year;

-- Q8: Avg Sharpe by category
SELECT category, ROUND(AVG(sharpe_ratio),2) as avg_sharpe
FROM scheme_performance GROUP BY category ORDER BY avg_sharpe DESC;

-- Q9: Funds with drawdown > -20%
SELECT scheme_name, max_drawdown_pct FROM scheme_performance
WHERE max_drawdown_pct < -20 ORDER BY max_drawdown_pct ASC;

-- Q10: Top 5 investors
SELECT investor_id, COUNT(*) as transactions, ROUND(SUM(amount_inr),2) as total_invested
FROM investor_transactions GROUP BY investor_id ORDER BY total_invested DESC LIMIT 5;
"""

with open(r'C:\Users\PRAHARSHA\mutual_fund_analytics\sql\queries.sql', 'w') as f:
    f.write(queries)

print("✅ queries.sql saved!")
```

```python
# Create schema.sql
schema = """
CREATE TABLE IF NOT EXISTS nav_history (
    amfi_code INTEGER,
    date TEXT,
    nav REAL
);

CREATE TABLE IF NOT EXISTS investor_transactions (
    investor_id TEXT,
    transaction_date TEXT,
    amfi_code INTEGER,
    transaction_type TEXT,
    amount_inr REAL,
    state TEXT,
    city TEXT,
    city_tier TEXT,
    age_group TEXT,
    gender TEXT,
    annual_income_lakh REAL,
    payment_mode TEXT,
    kyc_status TEXT
);

CREATE TABLE IF NOT EXISTS scheme_performance (
    amfi_code INTEGER,
    scheme_name TEXT,
    fund_house TEXT,
    category TEXT,
    plan TEXT,
    return_1yr_pct REAL,
    return_3yr_pct REAL,
    return_5yr_pct REAL,
    expense_ratio_pct REAL,
    sharpe_ratio REAL,
    aum_crore REAL
);
"""

with open(r'C:\Users\PRAHARSHA\mutual_fund_analytics\sql\schema.sql', 'w') as f:
    f.write(schema)

print("✅ schema.sql saved!")
```
